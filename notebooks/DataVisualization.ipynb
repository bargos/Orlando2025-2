```python
import geopandas as gpd
import os
# import rtree
# import pygeos
import mapclassify
import geobr
```

```python
gdf = geobr.read_municipality(code_muni='all', year=2020)
gdf.head()
```

```python
poa = gdf[gdf.name_muni=='Pindamonhangaba']
poa
```

```python
poa.crs
```

```python
poa.explore(
    color = 'gray',
    tooltip = 'name_muni',
    name = 'Pindamonhangaba',
    tiles = "CartoDB positron", #'OpenStreetMap'    #, "CartoDB positron", “CartoDB dark_matter"
)
```

```python
gdf_base = geobr.read_schools()
gdf_base.head()
```

```python
gdf_base.crs
```

```python
gdf_base_poa = gpd.sjoin(gdf_base, poa)
gdf_base_poa.head()
```

```python
gdf_base_poa.columns
```

```python
m = poa.explore(
    color = 'gray',
    tooltip = 'name_muni',
    name = 'Pindamonhangaba',
    tiles =   "CartoDB positron", #“CartoDB dark_matter"'OpenStreetMap'  
)
gdf_base_poa.explore(
    m = m,
    column='government_level',
    scheme='naturalbreaks',
    cmap='Spectral',
    names = 'Escolas',
    legend=True,
    legend_kwds=dict(caption='Legenda'),
    marker_kwds=dict(radius=2),
    height = 600,
    width = 600
)
```

```python
m.save(outfile='Escolas_Pinda.html')
```
